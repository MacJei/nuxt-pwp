Говоря о видах рекомендательных систем, обычно [выделют](/courses/recommendation-systems/introduction-to-recommendation-systems/) два основных типа таких моделей — <i title="collaborative filtering based models">рекомендательные системы, основанные на коллаборативной фильтрации</i> или системы, <i title="content based models">основанные на содержании</i>. При этом часто упускают ещё один крайне интересную разновидность рекомендательных систем, которая часто показывает резуальтаты, не уступающие популярным нынче нейронным сетям (https://www.zotero.org/ihun/items/itemKey/7TLXZPRX).

Фактически, эта разновидность рекомендательных систем представляет собой адаптацию графовых методов, давно используемых в других сферах, например в социальных науках, под новую задачу. Обычно эти методы применяются для 1) визуализации структуры социальных сетей, 2) нахождения самых важных её членов, например узлов с наибольшим количеством связей с другими узлами (говоря в других терминах — с наивысшей центральностью) или через которые проходит наибольшее количество кратчайших путей (такие узлы называются брокерами), 3) а также для выделения меньших подсообществ внутри основной исследуемой группы. В социальных науках, понятное дело, в роли узлов чаще всего выступают люди в сообществах или слова в текстах. В одном из исследований, в котором я принимал участие, в роли узлов выступали тэги, присвоенные постам в instagram, а связью между парами узлов устанавливалась в случае их наличия у одного и того же поста (https://www.zotero.org/ihun/items/itemKey/JIN73KAW).

Адаптация этих методов для задачи рекомендаций происходит довольно просто и естественно, если взять за узлы в сети принять товарные категории и проводять между ними связи (рёбра) в случае, если они встречаются среди покупок одного человека. Если какая-то пара категорий встречается более, чем у двух человек, то присваиваем связи между данной парой узлов больший вес, который в самом просто случае будет соответствовать количествую людей, у которых встречается эта пара категорий.

А далее — всё как раньше. Если необходимо сегментировать товарные категории — применяем <i title="community detection">алгоритмы выделения сообществ</i>, если хотим рекомендовать пользователю какую-то категорию — используём, например, алгоритм <i title="random walks">случайного блуждания</i> (https://www.zotero.org/ihun/items/itemKey/W6J9Q7P2), как делают в одноклассниках для рекомендаций групп (https://www.zotero.org/ihun/items/itemKey/PMYH7RXU).

Рассмотрим, как это работает на конкретной задаче выделения товарных сегментов и отнесения к ним пользователей. То есть мы хотим разбить наши товары на небольшие кучки на основе совместных покупок, а затем рекомендовать каждую кучку определённому пользователю. Сегодня займёмся первой частью этой задачи — сегментацией.

За исходные данные возьмём связи между категориями покупок, записанные в формате <i title="edge list">списка рёбер</i> — каждая строка будет представлять ребро графа и его вес, т.е. количество пользователей, у которых встречалась данная пара категорий.

    
<pre class="language-sql">
<code class="language-sql">
SELECT
    t1.grp AS cat1
    , t2.grp AS cat2
    , COUNT(*) AS weight
FROM filtered_bills t1
INNER JOIN filtered_bills t2
    ON t1.client = t2.client
    AND t1.grp <> t2.grp
GROUP BY t1.grp, t2.grp 
</code>

Теперь на основе этих данных создадим граф. Для работы с графами в python используются два основных пакета — [networkx](https://networkx.github.io/) и [python-igraph](https://igraph.org/python/). Первый написан на python, из-за чего он не слишком подходит для больших графов, но зато имеет понятный интерфейс и развитую инфраструктуру вспомогательных библиотек. Второй является обёрткой над сишной библиотекой, поэтому легче переваривает большие объёмы данных, но менее удобен для проведения исследований. Я выберу первый вариант.

```python
g = nx.from_pandas_edgelist(df_edges, source="cat1", target="cat2", edge_attr=["weight"])
g.number_of_nodes(), g.number_of_edges()
```

Тут нас поджидает проблема — почти каждая категория была хотя бы один раз куплена с любой другой. Это означает, что для кластеризации нам нужно или использовать алгоритмы, поддерживащие установку веса рёбер, или придётся тем или иным образом отсекать лишние связи. Уменьшение количеста ребёр также будет необходимо для визуализации полученного графа, поскольку отрисовка всех 2.28 млн ребер сделает картинку нечитаемой.

Однако, и тот и другой способ в моём случае приводили к образованию небольшого числа очень крупных сообществ и большого числа сообществ с небольшим количеством, а часто и одним членом. К тому же, если выбран путь удаления ребер, не так просто определить, сколько ребер нужно удалять. Я пробовал оставлять топ $n$ ребёр для каждого узла, пробовал удалять рёбра, вес которых менее $n$-ого процентиля, но к хорошим результатам это не приводило.

```python
from gensim.corpora import Dictionary
from gensim.models import Word2Vec, WordEmbeddingSimilarityIndex
from gensim.similarities import SoftCosineSimilarity, SparseTermSimilarityMatrix
```

```python
model = Word2Vec(users, size=50, window=2, workers=multiprocessing.cpu_count()-1)
```

```python
model.wv.similar_by_word("Наушники полноразмерные_low")
```

```python
termsim_index = WordEmbeddingSimilarityIndex(model.wv)
dictionary = Dictionary(users)
similarity_matrix = SparseTermSimilarityMatrix(termsim_index, dictionary)
```

```python
axis_labels = list(dictionary.values())
df_similarity_matrix = pd.DataFrame(
    data=similarity_matrix.matrix.toarray(),
    index=axis_labels,
    columns=axis_labels
)
```

```python
df_similarity_matrix.head(2)
```

```python
g_w2v = nx.from_pandas_adjacency(df_similarity_matrix)

for node in g_w2v.nodes():
    g_w2v.remove_edge(node, node)
```

```python
communities_w2v = algorithms.louvain(g_w2v, weight='weight', resolution=1)

print("Выделено сообществ: ", len(communities_w2v.communities))

cut_ratio_w2v = communities_w2v.cut_ratio()
print(f"Cut ratio: score={cut_ratio_w2v.score:.2f}, std={cut_ratio_w2v.std:.2f}. Lower is better.")

df_communities_w2v = pd.DataFrame.from_records([
    (len(c), ", ".join(c))
    for c in communities_w2v.communities
], columns=["cnt", "nodes"])
df_communities_w2v
```

```python
import networkx as nx
from itertools import combinations

G = nx.Graph()
for num, user in enumerate(users):
    print(num, end="\r")
    for i, j in combinations(user, 2):
        edge_data = G.get_edge_data(i, j)
        if not edge_data:
            current_weight = 0
        else:
            current_weight = G.get_edge_data(i, j)["weight"]
        G.add_edge(i, j, weight=current_weight + 1)
```

```python
import json
import pandas as pd
import os
import networkx as nx

from tqdm import tqdm_notebook
from loguru import logger
from df_loader import auto_read_sql as read_sql
from pathlib import Path
from sqlalchemy import create_engine, text

from gensim.models import Word2Vec

import json
import pandas as pd
import numpy as np
from pandas import read_sql
import os
import community as community_louvain
from cdlib import algorithms, ensemble
import cdlib
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns

from tqdm import tqdm_notebook
from loguru import logger
from pathlib import Path
from sqlalchemy import create_engine, text
from itertools import combinations

credentials = json.load(open("../../../../modules/config/credentials.json"))
connection_string = f'postgresql+psycopg2://{credentials["USER"]}:{credentials["PASSWORD"]}@{credentials["GREENPLUM_HOST"]}/{credentials["GREENPLUM_DBNAME"]}'
engine = create_engine(connection_string)
connection = engine.connect()

sql_drop_prepared_bills = """DROP TABLE IF EXISTS filtered_bills;"""

level = "mtg_map.txtlg" # для сбора данных на уровне групп
#level = "wgh1_map.txtmd" # для сбора данных на уровне wgh1
ntiles_num = 3 # первый раз делал при 5

sql_prepare_bills = f"""
SELECT
	{level} AS grp
    , business_partner AS client
    , sales_rev
INTO TEMP filtered_bills
FROM kernel.f_spss_bill_position bills
INNER JOIN ods_bwp.pzmaterial materials
    ON materials.zmaterial = bills.zmaterial
INNER JOIN ods_bwp.tzmatl_grp AS mtg_map
	ON mtg_map.zmatl_grp = materials.zmatl_grp
INNER JOIN ods_bwp.tzrp_wgh1 AS wgh1_map
    ON wgh1_map.zrp_wgh1 = materials.zrp_wgh1
INNER JOIN ods_bwp.tzrp_wgh2 AS wgh2_map
    ON wgh2_map.zrp_wgh2 = materials.zrp_wgh2
INNER JOIN ods_bwp.tzrp_wgh3 AS wgh3_map
    ON wgh3_map.zrp_wgh3 = materials.zrp_wgh3
INNER JOIN ods_bwp.tzrp_wgh4 AS wgh4_map
    ON wgh4_map.zrp_wgh4 = materials.zrp_wgh4
WHERE 1=1
	AND wgh1_map.langu = 'R'
	AND wgh2_map.langu = 'R'
	AND wgh3_map.langu = 'R'
	AND wgh4_map.langu = 'R'
	AND mtg_map.langu = 'R'
	AND wgh4_map.txtmd IN ('Аксессуары', 'Бытовая техника', 'Цифровая техника', 'Развлечения', 'Аудио/Видео')
	AND wgh3_map.txtmd NOT IN ('Подарочные карты', 'Услуги клиентам', 'Демо и подарки', 'Товары с АЕИ', 'Благотворительность', 'Подарки М.Видео', 'Подарочные карты', 'M.Price', 'Новогодние Акс-', 'Прочие платежи', 'Утилизация смешанная', 'ТЕЛЕКОММУНИКАЦИОННЫЕ ПРОДУКТЫ', 'Цифровой помощник Мобильная техника', 'Пакеты ПО Мобильная техника', 'Серты для Э-Ю Забота о доме и о себе', 'Серты для Э-Ю Кухня', 'Страхование для Э-Ю Мобильная техника', 'Серты для Э-Ю Кино и Звук', 'Защита', 'УСТАНОВКА БЫТОВОЙ ТЕХНИКИ 1', 'УСТАНОВКА БЫТОВОЙ ТЕХНИКИ', 'Серты для Э-Ю Мобильная техника', 'Серты для Э-Ю Домашний офис', 'Цифровой контент Мобильная техника', 'УСТАНОВКА БЫТОВОЙ ТЕХНИКИ 5')
	AND wgh1_map.txtmd NOT IN ('Пакеты')
    AND bills.date_sap < NOW()
    AND bills.date_sap > NOW() - INTERVAL '1 year'
    AND bills.sales_rev >= 0; -- удаляем возвраты. всё равно же купил, интерес был

-- считаем перекупами тех, кто находится в том 1% по покупкам
-- но купил менее, чем в 2-х разных группах

DELETE FROM filtered_bills
USING (
	SELECT client
	FROM (
		SELECT *
			, NTILE(100) OVER (ORDER BY cnt) AS pctile
		FROM (
			SELECT 
				client
				, COUNT(DISTINCT grp) AS grp_count
				, SUM(cnt) AS cnt
			FROM (
				SELECT
					client
					, grp
					, COUNT(*) AS cnt
				FROM filtered_bills
				GROUP BY client, grp
			) t
			GROUP BY client
		) t
	) t 
	WHERE
		(pctile >= 100 AND grp_count <= 2)
		OR cnt < 5
) users2delete
WHERE filtered_bills.client = users2delete.client;
"""

sql_download_docs = f"""
SELECT 
    client
    , STRING_AGG(grp, '||') AS doc
FROM (
    SELECT 
        client
        , grp || '_' ||
        CASE
            WHEN segm IN (1) THEN 'low'
            WHEN segm IN (2, 3) THEN 'middle'
            WHEN segm IN (4) THEN 'hight'
        END AS grp
    FROM (
        SELECT
            client
            , grp
            , NTILE(4) OVER (PARTITION BY grp ORDER BY sales_rev) segm
        FROM filtered_bills
    ) t
) t
GROUP BY client
"""
version = "v2"
connection.execute(sql_drop_prepared_bills)
connection.execute(text(sql_prepare_bills))
df_documents = pd.read_sql(sql=sql_download_docs, con=connection, index_col="client")
connection.execute(sql_drop_prepared_bills)
```

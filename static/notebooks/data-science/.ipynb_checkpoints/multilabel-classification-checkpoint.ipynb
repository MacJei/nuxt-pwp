## Введение

Существуют три основных вида классификации. Самый простой вид — это бинарная классификация, когда каждый объект нужно отнести к одному из двух классов. Классическим примерjv такого типа классификации является модель оттока, которая предсказывает вероятность несовершения пользователем целевого действия в течении определённого времени, после которого он считается «утёкшим». Соответственно, здесь всего два исхода — совершение действия за время t или невосершение такого действия. Ещё один случай такого рода классификаций — модель, определяющая эмоциональною тональность текстов, например, отзывов на товар. Отзыв может быть положительным или отрицательным.

Более сложная модификация этого метода — многоклассовая классификация, при которой требуется выбрать из более, чем двух классов. Например, мы можем усложнить классификатор тональности текстов, потребовав от модели, чтобы она относила текст не только к одному из двух полюсов эмоциональной тональности, но и допускала бы возможность нейтрального текста — это уже тринарная классификация. Или, например, работая в ритейле мы хотим построить рекоммендательную систему, предсказывающую категорию следующей покупки, причём количество таких категорий может достигать несколько сотен. Всё это примеры случаев, когда необходима многоклассовая классификация.

Если мы продолжим равивать пример с предсказанием категории следующей покупки, то заметим, что в одном чеке клиент может купить товары из нескольких категорий, и нам хорошо бы иметь это ввиду. В таком случае необходимо обратиться к мультилейбл классификации — это такой тип классификации, когда каждый объект выборки может быть одновременно отнесён к нескольким классам. Лейблов в данном контексте называются метки классов — например, названия категорий.

## Мультилейбл классификаторы

### Способы построения мультилейбл классификаторов

Существует два основных способа построения мультилейбл классификаторов — через трансформацию задачи и с использованием адаптивных методов. В первом случае задача решается через построение нескольких бинарных или многоклассовых классификаторов, а во втором задействуются алгоритмы, изначально рассчитанные именно на установление связей между объектом и множество лейблов.

Вторая группа алгоритмов велика и разнообразна, поэтому сейчас я остановлюсь на рассмотрении способов трансформации задачи.

#### Binary relevance

Техника под названием binary relevance — это наиболее интуитивный способ трансформации задачи мультилейбл классификации, при котором она трандля для каждого лейбла строится независимый бинарный классификатор. Таким образом большая, задача разбивается на q более простоых задач, где q — это количество лейблов.

Этот метод обычно используют в первую оцередь, когда выбирают подход, основанный на трансформации задачи, поскольку несмотря на свою простоту он показывает неплохие результаты. Однако, у него есть один существенный недостаток — поскольку бинарные классификаторы совершенно независимы друг от друга, они игнорируют возможную корреляцию лейблов, что может привести в неудовлетворительным результатам, когда эта корреляция присутствует.

Впрочем, эта же независимость каждого классификатора позволяет гибко настраивать каждый из них, индивидуально подбирая коэффициенты для ремемплинга, границу бинаризации и другие параметры.
